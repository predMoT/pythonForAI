```python
import pandas as pd
```

```python
df = pd.read_csv('HousingData.xls')
```

```python
print(df.head())
```

```python
print(df.isnull().sum())
```

```python
df.fillna(df.mean(), inplace = True)
```

```python
print(df.isnull().sum())
```

```python
# Veri setinin genel özetini inceleyelim
print(df.describe())
```

```python
# Veri setini 'boston_housing_ready.csv' olarak kaydedeceğiz
df.to_csv('boston_housing_ready.csv', index=False)
```

```python
print(df.columns)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Seaborn ile scatterplot çizme
sns.scatterplot(x='AGE', y='MEDV', data=df)

# Başlık ve etiketler ekleyelim
plt.title('AGE ile MEDV (Ev Fiyatları) Arasındaki İlişki')
plt.xlabel('AGE (Ortalama Yaş)')
plt.ylabel('MEDV (Ev Fiyatları)')

# Grafiği gösterelim
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# RM ile MEDV arasındaki ilişkiyi lineplot ile çizme
sns.scatterplot(x='RM', y='MEDV', data=df)

# Başlık ve etiketler ekleyelim
plt.title('RM ile MEDV (Ev Fiyatları) Arasındaki Eğilim')
plt.xlabel('RM (Ortalama Oda Sayısı)')
plt.ylabel('MEDV (Ev Fiyatları)')

# Grafiği gösterelim
plt.show()
```
